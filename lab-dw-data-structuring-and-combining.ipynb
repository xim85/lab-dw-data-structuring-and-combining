```python
import pandas as pd
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.columns = pd.Series(df.columns).apply(lambda col: col.lower().replace(' ', '_'))
```

```python
df.rename(columns={"st" : "state"},inplace=True)
df.head()
```

```python
df.isna().sum()
```

```python
df["number_of_open_complaints"] = df["number_of_open_complaints"].str.split("/").str[1]
```

```python
df["customer_lifetime_value"] = df["customer_lifetime_value"].str.replace("%","")
df.head()
```

```python
df.isna().sum()
```

```python
url_2 = ("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")
df_2 = pd.read_csv(url_2)
url_3 =("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")
df_3 = pd.read_csv(url_3)
```

```python
df_2.head()
```

```python
df_2.columns = pd.Series(df_2.columns).apply(lambda col: col.lower().replace(' ', '_'))
```

```python
df_2.rename(columns={"st" : "state"},inplace=True)
```

```python
df_2.isna().sum()
```

```python
df_2.head()
```

```python
df_2['customer_lifetime_value'] = df_2['customer_lifetime_value'].str.replace("%","")
df_2.head()
```

```python
df["number_of_open_complaints"] = df["number_of_open_complaints"].str.split("/").str[1]
```

```python
df_3.head()
```

```python
df_3.isna().sum()
```

```python
df_3.columns = pd.Series(df_3.columns).apply(lambda col: col.lower().replace(' ', '_'))
df_3.head()
```

```python
df_combined =pd.concat([df, df_2, df_3], axis=0, ignore_index=True)
```

```python
df_combined.shape
```

```python
df_combined.head()
```

```python
df_combined = df_combined.reset_index()
```

```python
df_combined.set_index("index", inplace =True)
```

```python
df_combined
```

```python
df_combined.isna().sum()
```

```python
df_combined.dropna(how="all", inplace= True)
```

```python
df_combined.gender.unique()
```

```python
gender_mapping = {"M": "M",
"F" : "F",
"Femal": "F", 
"Male": "M"}
df_combined.gender = df_combined.gender.map(gender_mapping)
df_combined.gender.unique()
```

```python
state_mapping = {"AZ": "Arizona", "WA": "Washington", "Cali": "California"}
df_combined["state"] = df_combined ["state"].replace(state_mapping)
df_combined.state.unique()
```

```python
education_mapping = {"Bachelors": "Bachelor",}
df_combined["education"] = df_combined["education"].replace(education_mapping)
```

```python
vehicle_class_mapping = {"Sports Car": "Luxury", "Luxury SUV": "Luxury", "Luxury Car": "Luxury"}
df_combined["vehicle_class"] = df_combined["vehicle_class"].replace(vehicle_class_mapping)
df_combined.vehicle_class.unique()
```

```python
df_combined.isna().sum()
```

```python
df_combined["gender"] = df_combined["gender"].fillna(df_combined["gender"].mode()[0])
df_combined.isnull().sum()
```

```python
df_combined["customer_lifetime_value"] = df_combined[["customer_lifetime_value"]].applymap(float)
df_combined.dtypes
```

```python
df_combined.isna().sum()
```

```python
df_combined.shape
```

```python
df_combined = df_combined.drop_duplicates()
df_combined.shape
```

```python
df_combined.isnull().sum()
```

```python
df_combined["customer_lifetime_value"] = df_combined["customer_lifetime_value"].fillna(df_combined["customer_lifetime_value"].mean(0))
```

```python
df_combined.isnull().sum()
```

```python
df_combined["number_of_open_complaints"] = df_combined["number_of_open_complaints"].fillna(df_combined["number_of_open_complaints"].mode()[0])
df_combined.isnull().sum()
```
